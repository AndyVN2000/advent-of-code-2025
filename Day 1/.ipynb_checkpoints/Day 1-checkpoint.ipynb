```python
def left(start, steps):
    result = start - steps
    if result < 0:
        return left(100, abs(result))
    else:
        return result

def right(start, steps):
    result = start + steps
    if result > 99:
        return right(0, result - 100)
    else:
        return result

def rotateDial(start, direction, steps):
    if direction == 'L':
        left(start, steps)
    elif direction == 'R':
        right(start, steps)
```

```python
currentNumber = 50
zero_hits = 0

with open('input.txt', 'r') as file:
    for line in file:
        direction = line[0]
        steps = int(line.strip()[1:])
        print(f'steps: {type(steps)}')
        print(f'currentNumber: {type(currentNumber)}')
        currentNumber = rotateDial(currentNumber, direction, steps)
        if currentNumber == 0:
            zero_hits += 1

f'The answer is {zero_hits}!'
```
