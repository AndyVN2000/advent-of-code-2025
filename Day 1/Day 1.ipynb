# Part One

```python
def left(start, steps):
    result = start - steps
    if result < 0:
        return left(100, abs(result))
    else:
        return result

def right(start, steps):
    result = start + steps
    if result > 99:
        return right(0, result - 100)
    else:
        return result

def rotateDial(start, direction, steps):
    if direction == 'L':
        return left(start, steps)
    elif direction == 'R':
        return right(start, steps)
```

```python
currentNumber = 50
zero_hits = 0

with open('input.txt', 'r') as file:
    for line in file:
        direction = line[0]
        steps = int(line.strip()[1:])
        currentNumber = rotateDial(currentNumber, direction, steps)
        if currentNumber == 0:
            zero_hits += 1

f'The answer is {zero_hits}!'
```

# Part Two

```python
def left(start, steps, wraps=0):
    result = start - steps
    if result < 0:
        if start == 0:
            return left(100, abs(result))
        return left(100, abs(result), wraps=wraps+1)
    else:
        return result, wraps

def right(start, steps, wraps=0):
    result = start + steps
    if result == 100:
        return 0, wraps
    if result > 99:
        return right(0, result - 100, wraps=wraps+1)
    else:
        return result, wraps

def rotateDial(start, direction, steps):
    if direction == 'L':
        return left(start, steps)
    elif direction == 'R':
        return right(start, steps)
```

```python
currentNumber = 50
zero_hits = 0

with open('input.txt', 'r') as file:
    for line in file:
        direction = line[0]
        steps = int(line.strip()[1:])
        currentNumber, wraps = rotateDial(currentNumber, direction, steps)
        zero_hits += wraps
        if currentNumber == 0:
            zero_hits += 1

f'The answer is {zero_hits}!'
```
