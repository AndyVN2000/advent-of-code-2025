# Part One

```python
def find_max_joltage(battery_bank):
    bat1_index = 0
    bat2_index = 0
    bat1_jolt = 0
    bat2_jolt = 0

    for i in range(0, len(battery_bank)):
        joltage = int(battery_bank[i])
        
        if joltage > bat1_jolt:
            bat1_index = i
            bat1_jolt = joltage

    for i in range(bat1_index + 1, len(battery_bank)):
        joltage = int(battery_bank[i])
        
        if joltage > bat2_jolt:
            bat2_index = i
            bat2_jolt = joltage
    
    return bat1_index, bat2_index, bat1_jolt, bat2_jolt

def compute_bank_joltage(battery1, battery2):
    return int(f'{battery1}{battery2}')
```

```python
max_bank_joltages = []

with open('input.txt', 'r') as file:
    for bank in file:
        bat1_index, bat2_index, bat1_jolt, bat2_jolt = find_max_joltage(bank.strip())
        bank_joltage = compute_bank_joltage(bat1_jolt, bat2_jolt)
        max_bank_joltages.append(bank_joltage)

solution = sum(max_bank_joltages)

f'The answer is {solution}!'
```
