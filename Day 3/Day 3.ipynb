# Part One

```python
def find_max_joltage(battery_bank):
    first_battery = 0
    second_battery = 0
    first_index = 0
    second_index = 0

    for i in range(0, len(battery_bank) - 1):
        joltage = int(battery_bank[i])
        if joltage > first_battery:
            first_battery = joltage
            first_index = i

    for i in range(first_index + 1, len(battery_bank)):
        joltage = int(battery_bank[i])
        if joltage > second_battery:
            second_battery = joltage
            second_index = i
    
    return first_battery, second_battery

def compute_bank_joltage(battery1, battery2):
    return int(f'{battery1}{battery2}')
```

```python
max_bank_joltages = []

with open('input.txt', 'r') as file:
    for bank in file:
        first_battery, second_battery = find_max_joltage(bank.strip())
        bank_joltage = compute_bank_joltage(first_battery, second_battery)
        max_bank_joltages.append(bank_joltage)

solution = sum(max_bank_joltages)

f'The answer is {solution}!'
```

```python
def find_max_joltage(battery_bank):
    batteries = [0] * 12
    battery_ctr = 0
    start_index = 0
    
    for i in range(11, -1, -1):
        limit = len(battery_bank) - i
        for j in range(start_index, limit):
            joltage = int(battery_bank[j])
            if joltage > batteries[battery_ctr]:
                batteries[battery_ctr] = joltage
                start_index = j

        battery_ctr += 1
        
    return batteries

def compute_bank_joltage(batteries):
    digits = ''
    for battery in batteries:
        digits = digits + str(battery)
    return int(digits)
```
