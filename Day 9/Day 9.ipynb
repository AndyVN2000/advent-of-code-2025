```python
with open ('input.txt', 'r') as file:
    lines = file.readlines()
    # List of tuples, representing coordinates in the tile floor.
    # Coordinates: (column, row)
    # ... how heretical...
    red_tiles = [ tuple([int(s) for s in l.strip().split(',')]) for l in lines]

# red_tiles
```

```python
largest_area = 0

for c0, r0 in red_tiles:
    for c1, r1 in red_tiles:
        width = abs(c0 - c1) + 1
        height = abs(r0 - r1) + 1
        area = width * height
        if area > largest_area:
            largest_area = area

largest_area
```
