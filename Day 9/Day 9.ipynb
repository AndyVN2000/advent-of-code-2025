# Part One

```python
with open ('input.txt', 'r') as file:
    lines = file.readlines()
    # List of tuples, representing coordinates in the tile floor.
    # Coordinates: (column, row)
    # ... how heretical...
    red_tiles = [ tuple([int(s) for s in l.strip().split(',')]) for l in lines]

#red_tiles
```

```python
largest_area = 0

def compute_area(p0, p1):
    c0, r0 = p0
    c1, r1 = p1
    width = abs(c0 - c1) + 1
    height = abs(r0 - r1) + 1
    return width * height

for p0 in red_tiles:
    for p1 in red_tiles:
        area = compute_area(p0, p1)
        if area > largest_area:
            largest_area = area

largest_area
```

# Part Two

```python
import matplotlib.pyplot as plt

x_axis = []
y_axis = []

for c, r in red_tiles:
    x_axis.append(c)
    y_axis.append(-r)

plt.plot(x_axis, y_axis, '-')
```

```python
from shapely import Point, Polygon

polygon = Polygon(red_tiles)

largest_area = 0

for p0 in red_tiles:
    c0, r0 = p0
    for p1 in red_tiles:
        c1, r1 = p1
        square = Polygon([(c0, r0), (c1, r0), (c1, r1), (c0, r1)])
        if not polygon.contains(square):
            continue
        area = compute_area(p0, p1)
        if area > largest_area:
            largest_area = area

largest_area
```
