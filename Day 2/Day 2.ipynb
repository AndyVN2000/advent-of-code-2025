# Part One

```python
def is_invalid_id(s):
    length = len(s)
    if length % 2 != 0:
        return False
    shift = int(length / 2)
    for i in range(0, shift):
        if s[i] != s[i+shift]:
            return False
    return True

def find_invalid_ids(start, end):
    result = []
    for i in range(start, end+1):
        if is_invalid_id(str(i)):
            result.append(i)
    return result
```

```python
with open('input.txt', 'r') as file:
    id_ranges = file.read().split(',')

total_invalids = []

for r in id_ranges:
    start, end = tuple(r.split('-'))
    invalid_ids = find_invalid_ids(int(start), int(end))
    total_invalids.extend(invalid_ids)

solution = sum(total_invalids)

f'The answer is {solution}!'
```

```python
def is_invalid_id(s):

    def check_shift(shift, length):
        for j in range(0, length-shift):
            #print(f's[j]: {s[j]}')
            #print(f's[j+shift]: {s[j+shift]}')
            #print(f'is not equal: {s[j] != s[j+shift]}')
            if s[j] != s[j+shift]:
                return False
        return True
    
    length = len(s)
    #print(f'length: {length}')
    shift_results = []
    for shift in range(1, int(length/2)+1):
        #print(f'shift: {shift}')
        if length % shift != 0:
            continue
        shift_finds_invalid = check_shift(shift, length)
        shift_results.append(shift_finds_invalid)
    
    return any(shift_results)

def find_invalid_ids(start, end):
    result = []
    for i in range(start, end+1):
        if is_invalid_id(str(i)):
            result.append(i)
    return result
```

```python
with open('input.txt', 'r') as file:
    id_ranges = file.read().split(',')

total_invalids = []

for r in id_ranges:
    start, end = tuple(r.split('-'))
    invalid_ids = find_invalid_ids(int(start), int(end))
    total_invalids.extend(invalid_ids)

solution = sum(total_invalids)

f'The answer is {solution}!'
```
