```python
def row_str_to_int(row):
    for i in range(0, len(row)):
        row[i] = int(row[i])
```

```python
with open('input.txt', 'r') as file:
    operands = []
    lines = file.readlines()
    for line in lines[:-1]:
        row = re.split("\s *", line.strip())
        row_str_to_int(row)
        operands.append(row)
        

    operators = re.split("\s *", lines[-1].strip())

np_operands = np.array(operands)

grand_total = 0

for col in range(0,len(operators)):
    operator = operators[col]
    result = None
    if operator == "+":
        result = np.sum(np_operands[:,col])
    elif operator == "*":
        result = np.prod(np_operands[:,col].astype(np.int64))
    else:
        raise Exception(f"operator is neither + or *\noperator: {operator}")
    grand_total += result

f'The answer is {grand_total}!'
```

# Part Two

```python
def is_digit(character):
    return bool(re.fullmatch(r'[0-9]', character))

def is_white_space(character):
    return bool(re.fullmatch(r'\s', character))

def is_operator(character):
    return bool(re.fullmatch(r'\+|\*', character))

def compute_equation(operands, operator):
    if operator == "+":
        return sum(operands)
    elif operator == "*":
        return np.prod(operands)
    else:
        raise Exception(f"operator is neither + or *\noperator: {operator}")
```

```python
is_operator('*')
```

```python
with open('input.txt', 'r') as file:
    lines = file.readlines()

for i in range(0, len(lines)):
    lines[i] = re.sub("\n","",lines[i])

# https://www.geeksforgeeks.org/python/python-convert-strings-to-character-matrix/
worksheet = [list(sub) for sub in lines]
worksheet[0][0]

row_count = len(worksheet)
col_count = len(worksheet[0])

grand_total = 0
operator = None

operands = []
solutions = []
for col in range(0, col_count):
    white_space_column_ctr = 0
    number = ''
    for row in range(0, row_count):
        character = worksheet[row][col]
        if is_digit(character):
            number = number + character
        elif is_white_space(character):
            white_space_column_ctr += 1
        elif is_operator(character):
            operator = character

        if white_space_column_ctr == row_count:
            solution = compute_equation(operands, operator)
            #print(f'operands: {operands}')
            #print(f'operator: {operator}')
            #print(f'solution: {solution}')
            solutions.append(solution)
            operator = None
            operands = []
    
    #print(f'number: {number}')
    if number != '':
        operands.append(int(number))
solution = compute_equation(operands, operator)
solutions.append(solution)
#print(f'operands: {operands}')
#print(f'operator: {operator}')
#print(f'solution: {solution}')


arr = np.array(solutions).astype(np.int64)
np.sum(arr)
```
