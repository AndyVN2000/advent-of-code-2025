```python
def to_matrix(file_lines):
    diagram = []
    
    for line in file_lines:
        row = list(line.strip())
        diagram.append(row)

    return diagram
```

```python
with open('example.txt', 'r') as file:
    lines = file.readlines()

diagram = to_matrix(lines)
diagram
```
